```python
# import my librairies
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import sqlite3
import gsw
from sklearn.linear_model import LinearRegression
```

```python
profile_speed = np.zeros((14,1))
ctcoeff = np.zeros((14,1))
```

```python
supercounter = 0

for files in np.arange(1,13,1):
    # SET METADATA FOR EACH TEST
    if files==1:
        filename = 'day2/208077_20211006_1123.rsk'
        numtest = 1
        indstart = [70290]
        indend = [118300]
        speed = [7.1]
    elif files ==2:
        filename = 'day2/208077_20211006_1318.rsk'
        numtest = 2
        indstart = [2682,125600]
        indend = [45580,175800]
        speed = [7.1,7.1]

    elif files ==3:
        filename = 'day2/208077_20211006_1504.rsk'
        numtest = 2
        indstart = [3701,126000]
        indend = [48860,164300]
        speed = [7.1,13.2]
                
    elif files ==4:
        filename = 'day2/208077_20211006_1601.rsk'
        numtest = 1
        indstart = [716]
        indend = [56980]
        speed = [18]
                
    elif files ==5:
        filename = 'day3/208077_20211007_1031.rsk'
        numtest = 1
        indstart = [73580]
        indend = [109880]
        speed = [18]
                
    elif files ==6:
        filename = 'day3/208077_20211007_1116.rsk'
        numtest = 1
        indstart = [14320]
        indend = [53190]
        speed = [13.2]
                
    elif files ==7:
        filename = 'day3/208077_20211007_1150.rsk'
        numtest = 1
        indstart = [10350]
        indend = [50405]
        speed = [23]   

    elif files ==8:
        filename = 'day3/208077_20211007_1232.rsk'
        numtest = 1
        indstart = [21940]
        indend = [60786]
        speed = [23]   
        
    elif files ==9:
        filename = 'day3/208077_20211007_1304.rsk'
        numtest = 1
        indstart = [8093]
        indend = [48135]
        speed = [30]   
        
    elif files ==10:
        filename = 'day3/208077_20211007_1342.rsk'
        numtest = 1
        indstart = [0]
        indend = [41255]
        speed = [30]   
        
    elif files ==11:
        filename = 'day3/208077_20211007_1425.rsk'
        numtest = 1
        indstart = [0]
        indend = [49955]
        speed = [42]   
        
    elif files ==12:
        filename = 'day3/208077_20211007_1507.rsk'
        numtest = 1
        indstart = [1433]
        indend = [43272]
        speed = [42]
        
        
    # Create a SQL connection to our SQLite database
    con = sqlite3.connect(filename)

    # query the data
    df = pd.read_sql_query("SELECT tstamp AS time, channel01 AS Cmeas, channel02 AS Tmeas, channel03 AS Pmeas, channel07 AS Tcond from data", con)
    con.close()

    # deal with time and date
    df['date'] = pd.to_datetime(df['time'],unit='ms')
    df = df.set_index('date')

    # compute secondary variables
    # computes approximate Sea Pressure
    df['Sea Pressure'] = df['Pmeas']-10.13
    # computes practical Salinity
    df['Smeas'] = gsw.SP_from_C(df['Cmeas'],df['Tmeas'],df['Sea Pressure'])
        
    for test in np.arange(0,numtest,1):

        # Crop the dataset
        rbr = df[indstart[test]:indend[test]]

        # Record flume speed
        profile_speed[supercounter] = speed[test]

        # compute normalized variables
        Tinit = np.nanmean(rbr.Tmeas[0:32*5])
        Tfinal = np.nanmean(rbr.Tmeas[-32*10:-1])
        Sinit = np.nanmean(rbr.Smeas[0:32*5])
        Sfinal = np.nanmean(rbr.Smeas[-32*10:-1])

        rbr['TN'] = 1- (rbr.Tmeas - Tinit)/(Tfinal - Tinit)
        rbr['SN'] = np.real(np.log(1-(rbr.Smeas - Sinit)/(Sfinal - Sinit)))

        # Compute elasped time
        t0 = np.where(rbr.TN<0.05)[0][0]
        rbr['etime'] = (rbr.time - rbr.time[t0])/1000
        
        # create figure and axis objects with subplots()
        fig,ax = plt.subplots(1,1,figsize=(15,10))
        # make a plot
        ax.plot(rbr.etime,rbr.TN, color="blue")
        # set x-axis label
        ax.set_xlabel("etime [s]",fontsize=14)
        # set y-axis label
        ax.set_ylabel("TN",color="b",fontsize=14)
        ax.grid()
        ax.set_title('Speed = '+str(profile_speed[supercounter][0]))

        # twin object for two different y-axis on the sample plot
        ax2=ax.twinx()
        # make a plot with different y-axis using second axis object
        ax2.plot(rbr.etime,rbr.SN,color="r")
        ax2.set_ylabel("SN",color="r",fontsize=14)

        # Compute CTCOEF
        
        # define reference salinity and conductivity a Salinity in last 10s
        Sref = Sfinal
        #Cref = gsw.C_from_SP(Sref*ones(size(Tmeas1)),Tmeas1,Pmeas1);

        # Compute Tanomaly
        rbr['Tanomaly'] = np.nan*rbr.Tmeas;
        blop = np.where(rbr.TN<0.05)[0];
        Tspace= np.arange(0,40,0.001)
        for ii in blop:
            Cspace = rbr.Cmeas[ii]*np.ones(Tspace.size);
            Sspace = gsw.SP_from_C(Cspace,Tspace,rbr['Sea Pressure'][ii]);
            I = np.argmin(np.abs(Sspace-Sref));
            rbr.Tanomaly[ii] = Tspace[I] - rbr.Tmeas[ii];

        rbr['Scor'] = gsw.SP_from_C(rbr.Cmeas,rbr.Tmeas+rbr.Tanomaly,rbr['Sea Pressure'])        

        Xtemp = np.array(rbr.Tcond[rbr.etime>90]-rbr.Tmeas[rbr.etime>90]);
        Ytemp = np.array(rbr.Tanomaly[rbr.etime>90]);
        mdl = LinearRegression().fit(Xtemp.reshape(-1,1), Ytemp.reshape(-1,1))
        ctcoeff[supercounter]= mdl.coef_[0][0]
        
        rbr['Tlong'] = ctcoeff[supercounter]*(rbr.Tcond-rbr.Tmeas)
        rbr['Scor'] = gsw.SP_from_C(rbr.Cmeas,rbr.Tmeas+rbr.Tlong,rbr['Sea Pressure'])
        rbr['SNcor'] = np.real(np.log(1-(rbr.Scor - Sinit)/(Sfinal - Sinit)))
        
        ax2.plot(rbr.etime,rbr.SNcor,color="g")
          
        # 
        Xtemp = np.array(rbr.Tcond[rbr.etime>90]-rbr.Tmeas[rbr.etime>90]);
        Ytemp = np.array(rbr.Tanomaly[rbr.etime>90]);
        mdl = LinearRegression().fit(Xtemp.reshape(-1,1), Ytemp.reshape(-1,1))
        ctcoeff[supercounter]= mdl.coef_[0][0]

        ax.plot(Xtemp,mdl.coef_[0][0]*Xtemp +mdl.intercept_[0],'r')
        ax.grid()
        ax.set_title('Speed = '+str(profile_speed[supercounter][0]))
        plt.savefig('SN_plunge'+str(supercounter+1)+'.png',bbox_inches='tight')
        close()

        # Ctcoef plot
        fig,ax = plt.subplots(1,1,figsize=(15,10))
        ax.plot(rbr.Tcond-rbr.Tmeas,rbr.Tanomaly, color="blue")
        ax.set_xlabel("∆T (Tcond - Tmeas) [°C]",fontsize=14)
        ax.set_ylabel("T_anomaly [°C]",color="b",fontsize=14)
        ax.set_title('Speed = '+str(profile_speed[supercounter][0]))        
        plt.savefig('Tanomaly_plunge'+str(supercounter+1)+'.png',bbox_inches='tight')
        close()
                          
        supercounter = supercounter + 1
        
```
